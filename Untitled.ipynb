```python
x=df.drop("booking_status",axis=1)
y=df["booking_status"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=40)
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
ypredict_train=knn.predict(x_train)
accuracy=accuracy_score(y_train,ypredict_train)
clf=confusion_matrix(y_train,ypredict_train)
report=classification_report(y_train,ypredict_train)
print(accuracy)
print(clf)
print(report)
```

```python
ypredict_test=knn.predict(x_test)
accuracy=accuracy_score(y_test,ypredict_test)
clf=confusion_matrix(y_test,ypredict_test)
report=classification_report(y_test,ypredict_test)
print(accuracy)
print(clf)
print(report)
```
